```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 5.Illustrate relationship between Total Bags and Small Bags using scatter plot.

```python
plt.scatter(data=dataset, x = 'Total Bags', y = 'Small Bags', s = 30, c = 'r')
plt.title('Relation between Total Bags and Smal Bags')
plt.xlabel('Total Bags')
plt.ylabel('Small Bags')
plt.show()
```

### 6.Construct AveragePrice distribution.

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 7.5,5
sns.distplot(dataset.AveragePrice, bins = 20)
```

### 7.Which type of Avocados are expensive?(with boxplot)

```python
sns.boxplot(x="type", y="AveragePrice", data = dataset, palette='dark')
```

### 8.Construct heatmap correlation matrix.

```python
f,ax=plt.subplots(figsize = (14,14))
sns.heatmap(dataset.corr(),annot= True,linewidths=0.5,fmt = ".2f")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.title('Correlation Show Map')
plt.savefig('correllation_graph.png')
plt.show()
```

### 9.Find relationship between large and small bags using scatterplot.(use jointplot function)

```python
sns.jointplot(x = "Large Bags", y = "Small Bags", data = dataset, kind = 'scatter')
```

### 10.Find linear relationship between Total Volume and Bags for each avocado type.(use lmplot function)

```python
sns.lmplot(y = "Total Volume", x = "Total Bags", data = dataset, col = "type", palette = "bright" )
```
